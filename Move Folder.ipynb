#### Original

```python
import os,shutil

folder = "Superstore_Sales"
folder_original = "Table 1 - Original"

src = fr"D:\Web Designing\Projects\{folder}\{folder_original}"
dst = fr"D:\Web Designing\Projects\{folder}\Table 3 - Final\1.Original"

def get_unique_path(dst_path):
    """If file exists, add _1, _2, etc."""
    if not os.path.exists(dst_path):
        return dst_path
    base, ext = os.path.splitext(dst_path)
    counter = 1
    while True:
        new_path = f"{base}_{counter}{ext}"
        if not os.path.exists(new_path):
            return new_path
        counter += 1
try:
    if not os.path.exists(src):
        print("Source folder not found")
    else:
        os.makedirs(dst, exist_ok=True)
        for item in os.listdir(src):
            src_path = os.path.join(src, item)
            if os.path.isfile(src_path):
                dst_path = os.path.join(dst, item)
                final_path = get_unique_path(dst_path)
                shutil.move(src_path, final_path)
        print("All files moved successfully")
except Exception as e:
    print("Error:", e)
```

#### 1. Find the number of orders in each region.

```python
query=""" 
select region,count(*) 
from train
group by region
order by count(*) desc
"""
cur.execute(query)
data=cur.fetchall()
columns = [col[0] for col in cur.description]
df = pd.DataFrame(data, columns=columns)
df = df.rename(columns={col: f"{col} (c{i+1})" for i, col in enumerate(df.columns)})
display(df.head().style.hide(axis="index").set_properties(**{'text-align': 'left'}).set_table_styles([{'selector': 'th', 'props': [('text-align', 'center')]},{'selector': 'th', 'props': [('border', '1.5px solid white')]},{'selector': 'td', 'props': [('border', '1px solid white')]}]))
```

#### 2. Find the top 5 cities with the highest total sales.

```python
query=""" 
with a as (
select city as c, sum(sales) as total_sales,
dense_rank() over(order by sales desc) as drank
from train
group by c,sales)
select c ,round(total_sales,2)
from a
where drank<=5
order by total_sales desc
"""
cur.execute(query)
data=cur.fetchall()
columns = [col[0] for col in cur.description]
df = pd.DataFrame(data, columns=columns)
df = df.rename(columns={col: f"{col} (c{i+1})" for i, col in enumerate(df.columns)})
display(df.head().style.hide(axis="index").set_properties(**{'text-align': 'left'}).set_table_styles([{'selector': 'th', 'props': [('text-align', 'center')]},{'selector': 'th', 'props': [('border', '1.5px solid white')]},{'selector': 'td', 'props': [('border', '1px solid white')]}]))
```

#### 3. Calculate monthly total sales using order_date.

```python
query=""" 
with a as (
select extract(month from order_date) as m, 
extract(year from order_date) as y,
sales as s
from train
)
select y,m,round(sum(s),2)
from a
group by y,m
order by y,m
"""
cur.execute(query)
data=cur.fetchall()
columns = [col[0] for col in cur.description]
df = pd.DataFrame(data, columns=columns)
df = df.rename(columns={col: f"{col} (c{i+1})" for i, col in enumerate(df.columns)})
display(df.head().style.hide(axis="index").set_properties(**{'text-align': 'left'}).set_table_styles([{'selector': 'th', 'props': [('text-align', 'center')]},{'selector': 'th', 'props': [('border', '1.5px solid white')]},{'selector': 'td', 'props': [('border', '1px solid white')]}]))
```

#### 4. Find the most used ship_mode and its count.

```python
query=""" 
select ship_mode,count(*) as smc

from train
group by ship_mode
order by smc desc
limit 1

"""
cur.execute(query)
data=cur.fetchall()
columns = [col[0] for col in cur.description]
df = pd.DataFrame(data, columns=columns)
df = df.rename(columns={col: f"{col} (c{i+1})" for i, col in enumerate(df.columns)})
display(df.head().style.hide(axis="index").set_properties(**{'text-align': 'left'}).set_table_styles([{'selector': 'th', 'props': [('text-align', 'center')]},{'selector': 'th', 'props': [('border', '1.5px solid white')]},{'selector': 'td', 'props': [('border', '1px solid white')]}]))
```

#### 5. Find the top 10 customers by total sales.

```python
query=""" 
with a as (
select customer_id as cid, 
customer_name as cn,
sales as s
from train
)
select cn,round(sum(s),2) as total_sales 
from a
group by cid,cn
order by total_sales desc
limit 10
"""
cur.execute(query)
data=cur.fetchall()
columns = [col[0] for col in cur.description]
df = pd.DataFrame(data, columns=columns)
df = df.rename(columns={col: f"{col} (c{i+1})" for i, col in enumerate(df.columns)})
display(df.head().style.hide(axis="index").set_properties(**{'text-align': 'left'}).set_table_styles([{'selector': 'th', 'props': [('text-align', 'center')]},{'selector': 'th', 'props': [('border', '1.5px solid white')]},{'selector': 'td', 'props': [('border', '1px solid white')]}]))
```

#### 6. Find customers who placed orders in more than 3 different months.

```python
query=""" 
select customer_name 
from train
group by customer_id,customer_name
having count(distinct date_format(order_date,'%m-%Y'))>3
"""
cur.execute(query)
data=cur.fetchall()
columns = [col[0] for col in cur.description]
df = pd.DataFrame(data, columns=columns)
df = df.rename(columns={col: f"{col} (c{i+1})" for i, col in enumerate(df.columns)})
display(df.head().style.hide(axis="index").set_properties(**{'text-align': 'left'}).set_table_styles([{'selector': 'th', 'props': [('text-align', 'center')]},{'selector': 'th', 'props': [('border', '1.5px solid white')]},{'selector': 'td', 'props': [('border', '1px solid white')]}]))
```

#### 7. Calculate the percentage contribution of each category to total sales.

```python
query=""" 
with a as (
select cat,sum(s) as ts
from trains
group by cat
)
select  
cat,round(100.0*ts/sum(ts) over(),2) as percentage
from a
"""
cur.execute(query)
data=cur.fetchall()
columns = [col[0] for col in cur.description]
df = pd.DataFrame(data, columns=columns)
df = df.rename(columns={col: f"{col} (c{i+1})" for i, col in enumerate(df.columns)})
display(df.head().style.hide(axis="index").set_properties(**{'text-align': 'left'}).set_table_styles([{'selector': 'th', 'props': [('text-align', 'center')]},{'selector': 'th', 'props': [('border', '1.5px solid white')]},{'selector': 'td', 'props': [('border', '1px solid white')]}]))
```

#### 8. Find customers who made purchases in consecutive months.

```python
query=""" 
with a as (
select cid ,c,
extract(year_month from od) as ym,
lag(extract(year_month from od)) over(partition by cid order by extract(year_month from od) asc) as pym
from trains)
select count(*)
from a 
where ym-pym=1
"""
cur.execute(query)
data=cur.fetchall()
columns = [col[0] for col in cur.description]
df = pd.DataFrame(data, columns=columns)
df = df.rename(columns={col: f"{col} (c{i+1})" for i, col in enumerate(df.columns)})
display(df.head().style.hide(axis="index").set_properties(**{'text-align': 'left'}).set_table_styles([{'selector': 'th', 'props': [('text-align', 'center')]},{'selector': 'th', 'props': [('border', '1.5px solid white')]},{'selector': 'td', 'props': [('border', '1px solid white')]}]))
```

#### 9. Calculate average shipping delay for each ship_mode and identify the fastest one.

```python
query="""
with a as ( 
select od,sd,sm,
avg(timestampdiff(day,od,sd)) over(partition by sm) as td
from trains
order by td
)
select distinct sm ,td
from a
order by td asc 
limit 1
"""
cur.execute(query)
data=cur.fetchall()
columns = [col[0] for col in cur.description]
df = pd.DataFrame(data, columns=columns)
df = df.rename(columns={col: f"{col} (c{i+1})" for i, col in enumerate(df.columns)})
display(df.head().style.hide(axis="index").set_properties(**{'text-align': 'left'}).set_table_styles([{'selector': 'th', 'props': [('text-align', 'center')]},{'selector': 'th', 'props': [('border', '1.5px solid white')]},{'selector': 'td', 'props': [('border', '1px solid white')]}]))
```

#### 10. For every state, find the sub-category with the highest total sales.

```python
query=""" 
with a as (
select sta,sc,sum(s) as ts,
row_number() over(partition by sta order by sum(s) desc) as drank
from trains
group by sta,sc
)
select sta,sc,round(ts,2)
from a
where drank=1
"""
cur.execute(query)
data=cur.fetchall()
columns = [col[0] for col in cur.description]
df = pd.DataFrame(data, columns=columns)
df = df.rename(columns={col: f"{col} (c{i+1})" for i, col in enumerate(df.columns)})
display(df.head().style.hide(axis="index").set_properties(**{'text-align': 'left'}).set_table_styles([{'selector': 'th', 'props': [('text-align', 'center')]},{'selector': 'th', 'props': [('border', '1.5px solid white')]},{'selector': 'td', 'props': [('border', '1px solid white')]}]))
```

Pyramid With For

```python
a=1
b=6
c=b+1

for i in range(a,b):
    for j in range(a,i+1):
        print("*",end=" ")
    print()
    if(i==4):
            break
for i in range(a,b):
    for j in range(a,c-i):
        print("*",end=" ")
    print()


for i in range(a,b):
    s=" "
    t=b-i
    print(s*t,end="")
    for j in range(a,i+1):
        print("","*",end="")
    print()

for i in range(a,b):
    s=" "
    t=a-1+i
    print(s*t,end="")
    for j in range(a,c-i):
        print("","*",end="")
    print()

```
